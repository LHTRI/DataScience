### Load data

```python
# Load dữ liệu
original_df = pd.read_csv('..\\Table_preprocess\\normal_original_data.csv', encoding='utf-8')
data_df = original_df.copy()
data_df
```

### Process data

- Biến đổi thời gian sang đơn vị chuẩn giây

```python
minute = data_df['Time'].apply(lambda x: int(x/1000))
second = data_df['Time'].apply(lambda x: int((x%1000)/10))
decimal_second = data_df['Time'].apply(lambda x: (x%10)/10)
data_df['Time'] = minute * 60 + second + decimal_second
data_df['Time']
```

- Điều chỉnh thời gian cho những con ngựa có thứ hạng bằng nhau: Tăng hoặc giảm thời gian của 1 trong 2 con ngựa

```python
data_df = data_df.sort_values(by=['race_id','KakuteiJyuni'])
list_race_id = sorted(data_df['race_id'].unique())
list_adjust_Time = np.array([])

for race_id in list_race_id:
    adjust_Time = data_df[data_df['race_id']==race_id]['Time'].shift(-1)
    list_adjust_Time = np.append(list_adjust_Time, adjust_Time)
    
data_df['temp'] = list_adjust_Time
```

```python
delta_time = 0.05
data_df.loc[data_df['temp'].isna(), 'temp'] = -1
data_df['check_equal'] = data_df['Time'] == data_df['temp']
data_df['adjust_Time'] = data_df['Time']
data_df.loc[data_df['check_equal'], 'adjust_Time'] = data_df['Time'] - delta_time
data_df['index'] = data_df.index
data_df = data_df.sort_values(by='index')
data_df['Time'] = data_df['adjust_Time']
data_df = data_df.drop(['temp', 'check_equal', 'adjust_Time', 'index'], axis=1)
```

```python
data_df[data_df['race_id']==1][['Time', 'KakuteiJyuni']]
```

- Tạo thêm các cột mới: top3


```python
data_df['top3'] = 0
condition = (data_df['KakuteiJyuni'] == 1) | (data_df['KakuteiJyuni'] == 2) | (data_df['KakuteiJyuni'] == 3)
data_df.loc[condition, 'top3'] = 1
# Check
data_df[['KakuteiJyuni', 'top3']]
```

- Tạo cột tốc độ

```python
data_df['speed'] = data_df['Kyori'] / data_df['Time'] * 3.6
data_df
```
