## Lấy dữ liệu tại bảng SE

```python
from collections import Counter
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

- Get data path

```python
import os.path as path
data_path =  path.abspath(path.join('' ,"../../data/historical_data"))
data_path = data_path + '\\SE.pickle.gz'
data_path
```

```python
import pandas as pd
SE_df = pd.read_pickle(data_path)
SE_df
```

### 1. Kiểm tra dữ liệu

```python
SE_df.info()
```

```python
# Select Columns
SE_selected_columns =  [
                        #Key
                        'id$Year',
                        'id$MonthDay',
                        'id$JyoCD',
                        'id$Kaiji',
                        'id$Nichiji',
                        'id$RaceNum',
                        #
                        'Umaban', 
                        'KettoNum',
                        'UmaKigoCD',
                        'SexCD',
                        'HinsyuCD',
                        'KeiroCD',
                        'Barei',
                        'TozaiCD',
                        'ChokyosiCode',
                        'BanusiCode',
                        'Futan',
                        'KisyuCode',
                        'MinaraiCD',
                        'BaTaijyu',
                        'ZogenFugo',
                        'ZogenSa',
                        'IJyoCD',
                        'KakuteiJyuni',
                        'Time',
                        'Odds',
                        'TimeDiff'
                       ]
SE_process_df = SE_df[SE_selected_columns]
SE_process_df
```

```python
# Kiểm tra thông tin dữ liệu các cột
SE_process_df.info()
```

```python
# Convert column id$JyoCD to int
SE_process_df['id$JyoCD'] = SE_process_df['id$JyoCD'].astype('int64')
SE_process_df['id$JyoCD'].unique()
```

- Kiểm tra các khóa chính ['id$Year', 'id$MonthDay', 'id$JyoCD', 'id$Kaiji', 'id$Nichiji', 'id$RaceNum', 'KettoNum']

```python
id_columns = ['id$Year', 'id$MonthDay', 'id$JyoCD', 'id$Kaiji', 'id$Nichiji', 'id$RaceNum', 'KettoNum']
check = SE_process_df.groupby(id_columns, as_index=False)[['Umaban']].count().rename(columns={'Umaban': 'count'})
check[check['count']>=2]
```

- <span style="color:blue">Nhận xét:
    - Có rất nhiều mã code bị trùng lặp, ta chỉ giữ lại record đầu tiên

```python
SE_process_df.drop_duplicates(subset=id_columns, keep='first', inplace=True)
SE_process_df
```

```python
# Convert Bataijyu
SE_process_df['BaTaijyu'] = SE_process_df['BaTaijyu'].astype('float64')
```

```python
SE_process_df[SE_process_df['BaTaijyu']==0]
```

```python
# Loại đi BaTaijyu = 0 vì ko dc xuất phát
SE_process_df = SE_process_df[SE_process_df['BaTaijyu']!=0]
SE_process_df
```

```python
# Loại đi các trường hợp KakuteiJyuni == 0
SE_process_df = SE_process_df[SE_process_df['KakuteiJyuni']!=0]
SE_process_df
```

```python
# Kiểm tra ZogenSa'=='   '
SE_process_df[SE_process_df['ZogenSa']=='    ']
```

```python
# Convert 
condition = (SE_process_df['ZogenSa']!='   ')
SE_process_df['New_ZogenSa'] = ''
SE_process_df.loc[condition, 'New_ZogenSa'] = SE_process_df.loc[condition, 'ZogenSa'].astype('int64')
SE_process_df.loc[SE_process_df['ZogenFugo']=='-', 'New_ZogenSa'] = -1 * SE_process_df['New_ZogenSa']
SE_process_df
```

```python
condition = SE_process_df['KettoNum'] == 2012101877
Counter(SE_process_df[SE_process_df['New_ZogenSa']=='']['KettoNum']).most_common()
```

```python
SE_process_df[SE_process_df['New_ZogenSa']==''][condition]
```

- <span style="color:blue">Nhận xét:
    - Đa số trường hợp thiếu ZogenSa là do lần đầu tham gia, giai đoạn đầu của mỗi con ngựa thường là cân nặng sẽ tăng, tuy nhiên ta sẽ đặt là 0 vì khởi đầu mọi con ngựa sẽ như nhau
   

```python
SE_process_df.loc[SE_process_df['New_ZogenSa']=='', 'New_ZogenSa'] = 0
SE_process_df['ZogenSa'] = SE_process_df['New_ZogenSa']
SE_process_df.loc[SE_process_df['ZogenSa']==999, 'ZogenSa'] = 0
SE_process_df.drop(['New_ZogenSa'], axis=1, inplace=True)
SE_process_df
```

```python
# Drop ZogenFugo
SE_process_df.drop(['ZogenFugo'], axis=1, inplace=True)
```

```python
SE_process_df['ZogenSa'].describe()
```

- Xử lý cho trường hợp khối lượng bất thường BaTaijyu==999

```python
SE_process_df[SE_process_df['BaTaijyu']==999]
```

```python
SE_process_df[SE_process_df['KettoNum']==2011101732]
```

- Chỉ có 1 trường hợp ta điều chỉnh BaTaijyu = 450

```python
# Trường hợp không thể cân thì lấy khối lượng trung bình(BaTaijyu==999)
condition = (SE_process_df['KettoNum']==2011101732) & (SE_process_df['BaTaijyu']==999)
SE_process_df.loc[condition, 'BaTaijyu'] = 450
```
