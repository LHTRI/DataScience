- Get data path

```python
import os.path as path
data_path =  path.abspath(path.join('' ,"../../data/historical_data"))
data_path = data_path + '\\CH.pickle.gz'
data_path
```

```python
CH_df = pd.read_pickle(data_path)
CH_df
```

- Kiểm tra khóa chính ChokyosiCode

```python
CH_df['ChokyosiCode'].describe()
```

```python
# Check key ChokyosiCode
len(CH_df['ChokyosiCode'].unique())
```

- <span style="color:blue">Nhận xét:
    - Có rất nhiều code bị trùng lặp, ta chỉ giữ lại 1 record

```python
CH_df.drop_duplicates(subset=['ChokyosiCode'], keep='first', inplace=True)
CH_df
```

```python
# Select Columns
CH_select_columns =  [
                        'ChokyosiCode',
                        'SexCD',
                        'TozaiCD',
                        'Syotai',
                        #'SaikinJyusyo$1$GradeCD',
                        #'SaikinJyusyo$2$GradeCD',
                        #'SaikinJyusyo$3$GradeCD',
                     ]
CH_process_df = CH_df[CH_select_columns]
CH_process_df
```

```python
CH_process_df.info()
```

- Đổi tên các cột trừ key

```python
for column in CH_select_columns:
    if column != 'ChokyosiCode':
        new_column = 'CH_' + column
        CH_process_df[new_column] = CH_process_df[column]
        CH_process_df.drop(column, axis=1, inplace=True)
CH_process_df
```

```python
# Ghi file
CH_process_df.to_csv('CH_process_df.csv', index=False)
```
