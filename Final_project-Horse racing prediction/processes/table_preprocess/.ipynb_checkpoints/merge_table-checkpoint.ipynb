## Merge các bảng và lưu bằng tên original_data.csv, nornal_original_data.csv
- Sử dụng 6 bảng: SE, RA, UM, KS, CH, HN

```python
import pandas as pd
import numpy as np
from collections import Counter
```

### 1. Load data

- RA table

```python
RA_df = pd.read_csv('RA_process_df.csv')
RA_df
```

```python
RA_df.info()
```

- SE table

```python
SE_df = pd.read_csv('SE_process_df.csv')
SE_df
```

```python
SE_df.info()
```

- UM table

```python
UM_df = pd.read_csv('UM_process_df.csv')
UM_df.head()
```

```python
UM_df.info()
```

- KS table

```python
KS_df = pd.read_csv('KS_process_df.csv')
KS_df.head()
```

```python
KS_df.info()
```

- CH table

```python
CH_df = pd.read_csv('CH_process_df.csv')
CH_df.head()
```

```python
CH_df.info()
```

- HN table

```python
HN_df = pd.read_csv('HN_process_df.csv')
HN_df.head()
```

```python
HN_df.info()
```

### 2. Merge UM và HN với key = HansyokuNum và Ketto3Info$1$HansyokuNum

```python
UM_df['HansyokuNum'] = UM_df['UM_Ketto3Info$1$HansyokuNum']
```

```python
# Merge 2 table
UM_HN_df = pd.merge(UM_df, HN_df, on=['HansyokuNum'], how='left')
UM_HN_df
```

### 3. Merge SE và UM_HN với key = KettoNum

```python
# Merge 2 table
UM_HN_SE_df = pd.merge(SE_df, UM_HN_df, on=['KettoNum'], how='left')
UM_HN_SE_df
```

### 3. Merge UM_HN_SE và KS với key = KisyuCode

```python
# Merge 2 table
UM_HN_SE_KS_df = pd.merge(UM_HN_SE_df, KS_df, on=['KisyuCode'], how='left')
UM_HN_SE_KS_df
```

### 4. Merge UM_HN_SE_KS và CH với key = ChokyosiCode

```python
# Merge 2 table
UM_HN_SE_KS_CH_df = pd.merge(UM_HN_SE_KS_df, CH_df, on=['ChokyosiCode'], how='left')
UM_HN_SE_KS_CH_df
```

```python
UM_HN_SE_KS_CH_df.info()
```

### 5. Merge UM_HN_SE_KS_CH và RA với key = ['id$Year', 'id$MonthDay', 'id$JyoCD', 'id$Kaiji', 'id$Nichiji', 'id$RaceNum']

```python
# Merge 2 table
UM_HN_SE_KS_CH_RA_df = pd.merge(RA_df, UM_HN_SE_KS_CH_df, on=['id$Year', 'id$MonthDay', 'id$JyoCD', 'id$Kaiji', 'id$Nichiji', 'id$RaceNum'],
                                how='left')
UM_HN_SE_KS_CH_RA_df
```

```python
UM_HN_SE_KS_CH_RA_df.info()
```

- <span style="color:blue">Nhận xét:
    - Một số dòng có giá trị Null do bảng HN không đủ dữ liệu, chỉ có 104 giá trị, ta có thể remove 

```python
UM_HN_SE_KS_CH_RA_df.dropna(inplace=True)
UM_HN_SE_KS_CH_RA_df.info()
```

```python
# Ghi dữ liệu ra file
UM_HN_SE_KS_CH_RA_df.to_csv('original_data.csv', index=False)
```
