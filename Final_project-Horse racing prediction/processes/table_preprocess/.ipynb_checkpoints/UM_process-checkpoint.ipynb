- <span style="color:blue">Nhận xét:
    - Có rất nhiều con ngựa bị trùng lặp mã số do update thông tin nhưng chưa xóa, để đơn giản ta chỉ giữ lại một dòng

```python
UM_df.drop_duplicates(subset=['KettoNum'], keep='first', inplace=True)
UM_df
```

```python
# Select Columns
UM_selected_columns =  [
                        'KettoNum',
                        'Bamei',
                        'ZaikyuFlag',
                        'SexCD',
                        'HinsyuCD',
                        'KeiroCD',
                        'Ketto3Info$1$HansyokuNum',
                        'Ketto3Info$1$Bamei',
                        'TozaiCD',
                        'ChokyosiCode',
                        'BreederCode',  #
                        'BanusiCode',
                        'RuikeiHonsyoHeiti',
                        'RuikeiHonsyoSyogai',
                        'RuikeiFukaHeichi',
                        'RuikeiFukaSyogai',
                        'RuikeiSyutokuHeichi',
                        'RuikeiSyutokuSyogai',
                        ]
UM_process_df = UM_df[UM_selected_columns]
UM_process_df
```

```python
UM_process_df.info()
```

```python
# Đổi tên các cột TozaiCD thành UM_TozaiCD
for column in UM_selected_columns:
    if column != 'KettoNum':
        new_column = 'UM_' + column
        UM_process_df[new_column] = UM_process_df[column]
        UM_process_df.drop(column, axis=1, inplace=True)
UM_process_df
```

```python
UM_process_df.info()
```

```python
UM_process_df.describe(include='all')
```

```python
# Ghi file
UM_process_df.to_csv('UM_process_df.csv', index=False)
```
