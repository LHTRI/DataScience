```python
KS_df.drop_duplicates(subset=['KisyuCode'], keep='first', inplace=True)
KS_df
```

```python
for c in KS_df.columns:
    print(c)
```

```python
# Select Columns
KS_select_columns =  [
                        'KisyuCode',
                        'SexCD',
                        'SikakuCD',
                        'MinaraiCD',
                        'TozaiCD',
                        'Syotai',
                        'ChokyosiCode',
                        #'SaikinJyusyo$1$GradeCD',
                        #'SaikinJyusyo$2$GradeCD',
                        #'SaikinJyusyo$3$GradeCD',
                        
                     ]
KS_process_df = KS_df[KS_select_columns]
KS_process_df
```

```python
KS_process_df.info()
```

```python
KS_process_df.info()
```

- Đổi tên các cột 

```python
# Đổi tên các cột TozaiCD thành UM_TozaiCD
for column in KS_select_columns:
    if column != 'KisyuCode':
        new_column = 'KS_' + column
        KS_process_df[new_column] = KS_process_df[column]
        KS_process_df.drop(column, axis=1, inplace=True)
KS_process_df
```

```python
# Ghi dữ liệu
KS_process_df.to_csv('KS_process_df.csv', index=False)
```
