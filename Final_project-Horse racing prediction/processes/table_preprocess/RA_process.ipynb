## Lấy dữ liệu tại bảng RA

```python
import pandas as pd
import numpy as np
from collections import Counter
import os
import sys
```

- Get data path

```python
import os.path as path
data_path =  path.abspath(path.join('' ,"../../data/historical_data"))
data_path = data_path + '\\RA.pickle.gz'
data_path
```

```python
RA_df = pd.read_pickle(data_path)
RA_df
```

### 1. Kiểm tra dữ liệu

```python
RA_df.info() 
```

```python
for c, d in zip(RA_df.columns, RA_df.dtypes):
    print(c, (30-len(c))*'  ',d)
```

- Kiểm tra các khóa chính

```python
Keys = ['id$Year',                   
        'id$MonthDay',                
        'id$JyoCD',                  
        'id$Kaiji',                   
        'id$Nichiji',                 
        'id$RaceNum'
       ]
RA_df.drop_duplicates(subset=Keys, keep='first', inplace=True)
RA_df
```

- <span style="color:blue">Nhận xét:
    - Không có key nào bị trùng lặp

- Xử lý Giải thưởng

```python
# Tổng hợp giải thưởng trước trận đấu
syokinAfter = [ 'Honsyokin$1',                
               'Honsyokin$2',               
            'Honsyokin$3',                
            'Honsyokin$4',
            'Honsyokin$5',
            'Fukasyokin$1',
            'Fukasyokin$2',
            'Fukasyokin$3'
         ]
syokinBefore = [ 'HonsyokinBefore$1',                
                 'HonsyokinBefore$2',               
                'HonsyokinBefore$3',                
                'HonsyokinBefore$4',
                'HonsyokinBefore$5',
                'FukasyokinBefore$1',
                'FukasyokinBefore$2',
                'FukasyokinBefore$3'
               ]

for after, before in zip(syokinAfter, syokinBefore):
    #print(after, before)
    RA_df.loc[RA_df[before]!=0, after] = RA_df[before]

# Tạo cột mới
RA_df['Honsyokin'] = RA_df['Honsyokin$1'] + RA_df['Honsyokin$2'] + RA_df['Honsyokin$3'] + RA_df['Honsyokin$4'] + RA_df['Honsyokin$5']
RA_df['Fukasyokin'] = RA_df['Fukasyokin$1'] + RA_df['Fukasyokin$2'] + RA_df['Fukasyokin$3']
```

```python
# Check
#RA_df[['Honsyokin$3', 'HonsyokinBefore$3']][RA_df['HonsyokinBefore$3']!=0]
RA_df['Honsyokin']
```

- Tạo dữ liệu tháng

```python
RA_df['Month'] = RA_df['id$MonthDay'].apply(lambda x: int(x/100))
RA_df[['Month', 'id$MonthDay']][RA_df['Month']==12]
```

- Tạo id_columns: race_id, Year

```python
RA_df.reset_index(drop=True, inplace=True)
RA_df['race_id'] = RA_df.index + 1
RA_df['Year'] = RA_df['id$Year']
```

```python
RA_selected_columns =  ['id$Year',
                                        'id$MonthDay',
                                        'id$JyoCD',
                                        'id$Kaiji',
                                        'id$Nichiji',
                                        'id$RaceNum',
                                        'race_id',
                                        'Year',
                                        'Month',
                                        'RaceInfo$TokuNum',
                                        'RaceInfo$Nkai',
                                        'GradeCD',
                                        'JyokenInfo$SyubetuCD',
                                        'JyokenInfo$KigoCD',
                                        'JyokenInfo$JyuryoCD',
                                        'Kyori', 
                                        'TrackCD',
                                        'TenkoBaba$TenkoCD',
                                        'TenkoBaba$SibaBabaCD',
                                        'TenkoBaba$DirtBabaCD',
                                        'Honsyokin',
                                        'Fukasyokin',
                                      ]
RA_process_df = RA_df[RA_selected_columns]
RA_process_df
```

```python
# Kiểm tra thông tin
RA_process_df.info()
```

```python
# Convert column id$JyoCD to int
RA_process_df['id$JyoCD'] = RA_process_df['id$JyoCD'].astype('int64')
RA_process_df['id$JyoCD'].unique()
```

```python
RA_process_df.describe(include='all')
```

```python
# Remove nan value
RA_process_df = RA_process_df.dropna()
RA_process_df
```

```python
# Kiểm tra lại thông tin
RA_process_df.info()
```

```python
# Ghi dữ liệu vào file
RA_process_df.to_csv('RA_process_df.csv', index=False)
```
