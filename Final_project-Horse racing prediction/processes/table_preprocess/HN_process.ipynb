- <span style="color:blue">Nhận xét:
    - Có vài con ngựa bị trùng lặp mã số ta sẽ chỉ giữ lại 1 record

```python
HN_df.drop_duplicates(subset=['HansyokuNum'], keep='first', inplace=True)
HN_df
```

```python
# Select Columns
HN_selected_columns =  [
                        'HansyokuNum',
                        #'KettoNum',
                        'Bamei',
                        'SexCD',
                        'HinsyuCD',
                        'KeiroCD',
                        'HansyokuMochiKubun',
                        'HansyokuFNum',
                        'HansyokuMNum',
                        ]
HN_process_df = HN_df[HN_selected_columns]
HN_process_df
```

```python
HN_process_df.info()
```

```python
# Đổi tên các cột 
for column in HN_selected_columns:
    if column != 'HansyokuNum':
        new_column = 'HN_' + column
        HN_process_df[new_column] = HN_process_df[column]
        HN_process_df.drop(column, axis=1, inplace=True)
HN_process_df
```

```python
# Ghi File
HN_process_df.to_csv('HN_process_df.csv', index=False)
```
