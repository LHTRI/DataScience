## 2. Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

def train_model(X_train, y_train, y_train_df): 
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)
    return lr_model
```

```python
def evaluate(model, X, y_df):
    rate, r2, mae, rmse, result_df = predict(model, X, y_df)
    print('Sai số rmse:                    %.3f'%rmse)
    print('Hệ số xác định r2-score: %.3f'%r2)
    print('Tỉ lệ True positive:           %.3f'%rate)
    return result_df
```

- Train model

```python
model = train_model(X_train, y_train, y_train_df)
```

- Evaluate 

```python
# On train 
train_result_df = evaluate(model, X_train, y_train_df)
train_result_df
```

```python
# On test
test_result_df = evaluate(model, X_test, y_test_df)
test_result_df
```

## 3. Train XGBRegressor 

```python
from xgboost import XGBRegressor 
import xgboost as xg
```

```python
def Xg_model(X, y): 
    X_model = XGBRegressor(objective ='reg:linear',  n_estimators=10, seed=123) 
    X_model.fit(X, y)
    y_pred = X_model.predict(X)
    r2 = r2_score(y, y_pred)
    rmse = math.sqrt(mean_squared_error(y, y_pred))
    print('r2', r2)
    print('rmse', rmse)
    return X_model

X_model = Xg_model(X_train, y_train)
```

```python
# Predict
result_df = evaluate(X_model, X_test, y_test_df)
```
