- Tương quan độ tuổi với sức mạnh của ngựa

```python
# Tạo hàm vẽ đồ thị trendline
def myline(df, x ='Barei', y ='speed', deg=1, figsize=(12,6), legend_loc='upper right'):
    
    df_test = df[[x, y]]
    df_group = df_test.groupby(x, as_index=False).mean()
    X = [i for i in range(0, len(df_group))]
    Y = df_group[y]
    trend = np.polyfit(X, Y, deg)
    trend_line = np.poly1d(trend) 
    plt.figure(figsize=figsize);
    ax = sns.lineplot(x=x, y=y, data=df_group, marker='s', label=y);
    plt.plot(df_group[x], trend_line(X), label='trend line');
    plt.title('Line chart of speed by {}'.format(x));
    plt.legend(loc=legend_loc)
    return df_group, ax
```

```python
sns.set(font_scale=1.2, style='darkgrid')
myline(df_horse, x ='Barei', y ='speed', deg=3, figsize=(10,6));
plt.xticks(np.arange(14));
```

- Phân tích ANOVA

```python
from scipy import stats
import seaborn as sns

sns.set(font_scale=1.2, style='dark')

df_F_test = df_horse.groupby('Barei', as_index=False)

F_val = np.zeros([6, 6])
P_val = np.zeros([6, 6])

for i in range(6):
    for j in range(6):
        f_val, p_val = stats.f_oneway(df_F_test.get_group(i+2)['speed'],
                                      df_F_test.get_group(j+2)['speed'])  
        F_val[i,j] = f_val
        P_val[i,j] = p_val
                
df_F_val = pd.DataFrame(F_val, index=[2, 3, 4, 5, 6, 7])
df_P_val = pd.DataFrame(P_val, index=[2, 3, 4, 5, 6, 7])

fig, axes = plt.subplots(1, 2, figsize=(18,7))
fig_1 = sns.heatmap(df_F_val, annot=True, annot_kws={"size": 12},
                    xticklabels=[2, 3, 4, 5, 6, 7], ax=axes[0], linewidths=2)
fig_1.set_title('F value');
fig_2 = sns.heatmap(df_P_val, annot=True, annot_kws={"size": 12},
                    xticklabels=[2, 3, 4, 5, 6, 7], ax=axes[1], linewidths=2)
fig_2.set_title('p value');
fig.suptitle('Heat map of F-test value of difference of speed between age', fontsize=15);

fig.add_subplot(111, frame_on=False);
plt.tick_params(labelcolor="none", bottom=False, left=False);
plt.xlabel("Age");
plt.ylabel("Age");
plt.subplots_adjust(wspace=0.1);
```

- <span style="color:blue">Nhận xét:
    - Độ tuổi càng lớn thì tốc độ càng giảm, đồng nghĩa với xếp hạng càng thấp(vị trí gần cuối 1 cuộc đua). Tốc độ lớn nhất tại khoảng 4 tuổi và 6 tuổi tốc độ bắt đầu giảm dần

- Phân tích ANOVA

```python
from scipy import stats
import seaborn as sns

sns.set(font_scale=1.2, style='dark')

df_F_test = df_horse.groupby('SexCD', as_index=False)

F_val = np.zeros([3, 3])
P_val = np.zeros([3, 3])

for i in range(3):
    for j in range(3):
        f_val, p_val = stats.f_oneway(df_F_test.get_group(i+1)['speed'],
                                      df_F_test.get_group(j+1)['speed'])  
        F_val[i, j] = f_val
        P_val[i, j] = p_val
        
index = ['Male', 'Female', 'Castrated']     
df_F_val = pd.DataFrame(F_val, index=index)
df_P_val = pd.DataFrame(P_val, index=index)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig_1 = sns.heatmap(df_F_val, annot=True, annot_kws={"size": 15},
                    xticklabels=index, ax=axes[0], linewidths=2)
fig_1.set_title('F value');
fig_2 = sns.heatmap(df_P_val, annot=True, annot_kws={"size": 15},
                    xticklabels=index, ax=axes[1], linewidths=2)
fig_2.set_title('p value');
fig.suptitle('Heat map of F-test value of difference of speed between SexCD', fontsize=18);

fig.add_subplot(111, frame_on=False);
plt.tick_params(labelcolor="none", bottom=False, left=False);
plt.xlabel("SexCD");
plt.ylabel("SexCD");
plt.subplots_adjust(wspace=0.1);
```
