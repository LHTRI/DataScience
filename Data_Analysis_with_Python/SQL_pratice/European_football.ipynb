## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
# write your query
result = pd.read_sql("""SELECT Match.id, Match.season, Match.stage, Match.match_api_id, Match.home_team_api_id,
                               Match.away_team_api_id, Match.home_team_goal, Match.away_team_goal,
                               League.name AS "League_name", League.id AS "League_id",
                               Country.name AS "Country", Country.id AS "Country_id"
                          FROM Match
                          JOIN League 
                            ON Match.League_id = League.id
                          JOIN Country 
                            ON Match.country_id = Country.id;""" , conn)
result
```

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh UNION


```python
# write your query
result = pd.read_sql("""SELECT Team.team_long_name,
                               SUM(Match.goal) AS goal_total
                          FROM 
                               (SELECT Match.home_team_api_id AS team_api_id, Match.home_team_goal AS goal
                                  FROM Match
                                 UNION ALL
                                SELECT Match.away_team_api_id AS team_api_id, Match.away_team_goal AS goal
                                  FROM Match) AS Match
                          JOIN Team
                            ON Team.team_api_id = Match.team_api_id
                         GROUP BY Match.team_api_id
                         ORDER BY goal_total DESC;""",conn)
result
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
# write your query
result = pd.read_sql("""SELECT Team.team_api_id, Team.team_long_name,
                               COUNT() AS "match_total"  
                          FROM Match
                          JOIN Team
                            ON (Team.team_api_id = Match.home_team_api_id) OR (Team.team_api_id = Match.away_team_api_id)
                         GROUP BY Team.team_api_id
                         ORDER BY COUNT() DESC;""",conn)
                         
result      
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu đội nhà có bàn thắng > bàn thắng của đội khách, đội nhà là "giành chiến thắng" trong trận đấu
- Nếu đội nhà có số bàn thắng < bàn thắng của đội khách, đội nhà bị "thua" trong trận đấu này
- Nếu độ nhà có bàn thắng = bàn thắng của đội khách, đội là "hòa" trong trận đấu này
- Sử dụng câu lệnh "case when"


```python
# write your query
result = pd.read_sql("""SELECT Team.team_long_name,
                               COUNT(CASE WHEN goal_difference > 0 THEN 1 ELSE NULL END) AS win_match_total,
                               COUNT(CASE WHEN goal_difference < 0 THEN 1 ELSE NULL END) AS lose_match_total,
                               COUNT(CASE WHEN goal_difference = 0 THEN 1 ELSE NULL END) AS draw_match_total,
                               COUNT() AS match_total
                          FROM 
                               (SELECT Match.home_team_api_id AS team_api_id, (Match.home_team_goal - Match.away_team_goal) AS goal_difference 
                                  FROM Match
                                 UNION ALL
                                SELECT Match.away_team_api_id AS team_api_id, (Match.away_team_goal - Match.home_team_goal) AS goal_difference
                                  FROM Match) AS Match
                          JOIN Team
                            ON Team.team_api_id = Match.team_api_id
                         GROUP BY Match.team_api_id;""",conn)
result
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
# write your query
result = pd.read_sql("""SELECT Team.team_long_name,
                               COUNT(CASE WHEN goal_difference > 0 THEN 1 ELSE NULL END) AS win_match_total
                          FROM 
                               (SELECT Match.home_team_api_id AS team_api_id, (Match.home_team_goal - Match.away_team_goal) AS goal_difference 
                                  FROM Match
                                 UNION ALL
                                SELECT Match.away_team_api_id AS team_api_id, (Match.away_team_goal - Match.home_team_goal) AS goal_difference
                                  FROM Match) AS Match
                          JOIN Team
                            ON Team.team_api_id = Match.team_api_id
                         GROUP BY Match.team_api_id
                         ORDER BY win_match_total DESC
                         LIMIT 10;""",conn)
result
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
# write your query
result = pd.read_sql("""SELECT League.name AS "league_name", COUNT() AS league_match_total,
                               SUM(COUNT()) OVER() AS match_total,
                               ROUND(CAST(COUNT() AS FLOAT)/CAST(SUM(COUNT()) OVER() AS FLOAT)*100,2) AS "rate(%)"
                          FROM Match
                          JOIN League ON Match.League_id = League.id
                         GROUP BY League_id
                         ORDER BY COUNT() DESC;""",conn)
                         
result
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
# write your query
result = pd.read_sql("""SELECT League.name AS "league_name", SUM(Match.home_team_goal+Match.away_team_goal) AS league_goal_total,
                               SUM(SUM(Match.home_team_goal+Match.away_team_goal)) OVER() AS goal_total,
                               CAST(SUM(Match.home_team_goal+Match.away_team_goal) AS FLOAT)
                               /CAST(SUM(SUM(Match.home_team_goal+Match.away_team_goal)) OVER() AS FLOAT)*100 AS "rate(%)"
                          FROM Match
                          JOIN League ON Match.League_id = League.id
                         GROUP BY League_id
                         ORDER BY COUNT() DESC;""",conn)
                         
result
```

## Câu hỏi 19: Liệt kê thông tin cầu thủ:

- Quy đổi cân nặng sang kilogram
- Quy đổi chiều cao sang mét
- Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)

```python
# write your query
result = pd.read_sql("""SELECT player_api_id,
                               player_name,
                               birthday,
                               ROUND(height/100,2) AS "height(m)",
                               ROUND((weight*0.453592),2) AS "weight(kg)",
                               ROUND(weight* 0.453592/ (height*height/10000),2) AS bmi,
                               strftime('%Y',date()) - strftime('%Y',birthday) AS "age"
                          FROM Player;""",conn)

result
```

## Câu hỏi 20: Liệt kê cầu thủ lớn tuổi nhất

```python
# write your query
result = pd.read_sql("""SELECT player_api_id,
                               player_name,
                               birthday,
                               ROUND(height/100,2) AS "height(m)",
                               ROUND((weight*0.453592),2) AS "weight(kg)",
                               ROUND(weight* 0.453592/ (height*height/10000),2) AS bmi,
                               strftime('%Y',date()) - strftime('%Y',birthday) AS "age"
                          FROM Player
                         WHERE birthday = (SELECT MIN(birthday) FROM Player);""",conn)

result
```

## Câu hỏi 21: Liệt kê  cầu thủ có số trận đấu nhiều nhất

```python
# write your query
result = pd.read_sql("""SELECT Player_Attributes.player_api_id,
                               player_name,
                               ROUND(height/100,2) AS "height(m)",
                               ROUND((weight*0.453592),2) AS "weight(kg)",
                               ROUND(weight* 0.453592/ (height*height/10000),2) AS bmi,
                               strftime('%Y',date()) - strftime('%Y',birthday) AS "age",
                               number_match
                          FROM 
                               (SELECT player_api_id,
                                       COUNT() AS number_match,
                                       MAX(COUNT()) OVER() AS max_number_match
                                  FROM Player_Attributes
                                 GROUP BY player_api_id) AS Player_Attributes
                          JOIN Player
                            ON Player.player_api_id = Player_Attributes.player_api_id
                         WHERE number_match = max_number_match;""",conn)

result
```

## Câu hỏi 22: Liệt kê các cầu thủ có tổng số điểm đánh giá lớn hơn 80

```python
# write your query
result = pd.read_sql("""SELECT Player_Attributes.player_api_id,
                               player_name,
                               ROUND(height/100,2) AS "height(m)",
                               ROUND((weight*0.453592),2) AS "weight(kg)",
                               ROUND(weight* 0.453592/ (height*height/10000),2) AS bmi,
                               strftime('%Y',date()) - strftime('%Y',birthday) AS "age",
                               overall_rating
                          FROM Player_Attributes
                          JOIN Player 
                            ON Player.player_api_id = Player_Attributes.player_api_id
                         WHERE overall_rating > 80
                         GROUP BY Player.player_api_id
                         ORDER BY overall_rating DESC;""",conn)

result
```
